```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
latvju_muzika = pd.read_excel('Book1_latvju_muzika_jy.xlsx')


columns = ['DZVBA/SKUMS', 'NVE/BEIGAS', 'RAKSTS', 'ATMIA', 'BALSS', 'CELTNE',
       'CE', 'DABA', 'DARBS', 'DEFORMCIJA', 'EMOCIJAS', 'FIZIKA',
       'JAUNS/VECS', 'LIELS/MAZS', 'DAUDZ/MAZ', 'KAR', 'KOPANA', 'ERMENIS',
       'LAIKS', 'MRJUMI', 'MET/POTISKA', 'MET/MZIKA', 'PAAUDZES',
       'POLITIKA', 'PROCESS', 'PROFESIJAS', 'PS. PABAS', 'RELIIJA',
       'RITULS', 'ROBEAS', 'SAITES', 'SEN/DZIRDE', 'SEN/GARA', 'SEN/REDZE',
       'SEN/SMARA', 'SEN/TAUSTE', 'SKAA/KLUSUMS', 'SLPANS',
       'SOC/HIERARHIJA', 'SOC/KOPIENA', 'SOC/VARA', 'SPIRITULS', 'STATUSS',
       'STIPRUMS/SPKS', 'TEHNOLOIJA', 'TEKSTILS', 'TELPA', 'TRBA',
       'TRADCIJAS', 'TVERTNE', 'VALSTS/TAUTA', 'VESELBA/SLIMBA', 'VRTBA',
       'VIRZIENS', 'IDIOMA', 'FRAZEOLOISMS', 'PERSONIFIKCIJA', 'REFERENCE',
       'EIFMISMS', 'SVEVALODA', 'IRONIJA', 'METONMIJA', 'DIRIENTS',
       'DZIESMU SVTKI', 'ANSAMBLIS', 'KORIS', 'KLAUSTJS', 'MZIIS',
       'KRITIIS', 'KOMPONISTS', 'MZIKA', 'POSITIVE', 'NEGATIVE', 'hard_to_say', 'na']
DABA = {}
for col in columns:
    DABA[col] = np.sum(latvju_muzika.DABA[latvju_muzika[col] == 1])
DABA.pop('DABA', None)
plt.figure(figsize=(20,5))
plt.bar(range(len(DABA)), DABA.values(), align='center')
plt.xticks(range(len(DABA)), DABA.keys(), rotation=90)
plt.title("korelcijas ar dabas kategoriju")
plt.legend(loc='best')
#plt.show()
plt.savefig('korealcijas ar dabas kategoriju.png',bbox_inches="tight")
```
